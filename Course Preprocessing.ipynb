```python
d = []
for i in range(len(courses['Difficulty'])):
    if(courses['Difficulty'][i] == 'Easy'):
            d.append(0)
    elif(courses['Difficulty'][i] == 'Medium'):
        d.append(1)
    else:
        d.append(2)
```
